```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Load your Shorts dataset
df = pd.read_csv("C:/mysqldata/youtube_shorts.csv", encoding='latin1')
  # Clean column names

# Convert Upload Date & Views
df['Upload Date'] = pd.to_datetime(df['Upload Date'], dayfirst=True, errors='coerce')
df['Views'] = pd.to_numeric(df['Views'], errors='coerce')

# Prepare data for Prophet
prophet_df = df[['Upload Date', 'Views']].dropna()
prophet_df.rename(columns={'Upload Date': 'ds', 'Views': 'y'}, inplace=True)

# Fit Prophet model
model = Prophet()
model.fit(prophet_df)

# Forecast future
future = model.make_future_dataframe(periods=1,freq = 'M')
forecast = model.predict(future)

# Plot forecast
model.plot(forecast)
plt.title("30-Day Forecast of Views (YouTube Shorts)")
plt.show()
model.plot_components(forecast)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate
csv= pd.read_csv(r"C:\Users\Dell\Downloads\Team 12 Youtube Channel Shorts Performance Analysis Dataset.csv")

swot = {
    "Strengths": [
        "High view counts on several Shorts (up to 15K)",
        "Consistent posting frequency",
        "Strong niche content (Indian food)",
        "Good like-to-view ratio on many videos"
    ],
    "Weaknesses": [
        "Low comment engagement (few viewer replies)",
        "No use of hashtags to boost discoverability",
        "No call-to-action in Shorts to convert viewers to subscribers"
    ],
    "Opportunities": [
        "Use trending hashtags to reach wider audiences",
        "Add hooks and CTAs to encourage comments and subscriptions",
        "Repurpose Shorts for Instagram Reels and YouTube Stories",
        "Leverage viral videos for spin-off content"
    ],
    "Threats": [
        "YouTube algorithm volatility may drop Shorts visibility",
        "Increased competition in food Shorts niche",
        "Fatigue from similar content style or pacing",
        "Short shelf-life of virality"
    ]
}

swot_matrix = [
    [", ".join(swot["Strengths"]), ", ".join(swot["Weaknesses"])],
    [", ".join(swot["Opportunities"]), ", ".join(swot["Threats"])]
]

headers = ["Strengths", "Weaknesses"]
print("YouTube Shorts SWOT Analysis")
print(tabulate(swot_matrix, headers=headers, tablefmt="fancy_grid"))

swot_counts = {category: len(points) for category, points in swot.items()}

# Pie chart
plt.figure(figsize=(7, 7))
plt.pie(
    swot_counts.values(),
    labels=swot_counts.keys(),
    autopct='%1.1f%%',
    startangle=140,
    colors=["#2ecc71", "#e74c3c", "#3498db", "#e67e22"],
    textprops={'fontsize': 14}  # 👈 Increase font size here
)

plt.title("SWOT Analysis Distribution: YouTube Shorts")
plt.axis('equal')
plt.tight_layout()
plt.show
```
